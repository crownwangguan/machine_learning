```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
columns = ['user_id', 'item_id', 'rating', 'timestamp']
frame = pd.read_csv('ml-100k/u.data', sep='\t', names=columns)
frame.head()
```

```python
columns = ['item_id', 'movie_title', 'release_date', 'video_release_date', 'IMDB_URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Fil-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=columns, encoding='latin-1')
movies.head()
```

```python
combined_movies_data = pd.merge(frame, movies, on='item_id')
combined_movies_data.head()
```

```python
combined_movies_data.groupby('item_id')['rating'].count().sort_values(ascending=False).head()
```

```python
Filter = combined_movies_data['item_id'] == 50
combined_movies_data[Filter]['movie_title'].unique()
```

```python
rating_crosstab = combined_movies_data.pivot_table(values='rating', index='user_id', columns='movie_title', fill_value=0)
rating_crosstab.head(10)
```

```python
rating_crosstab.shape
```

```python
X = rating_crosstab.values.T
```

```python
X.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=17)
```

```python
resultant_matrix = SVD.fit_transform(X)
```

```python
resultant_matrix.shape
```

```python
corr_mat = np.corrcoef(resultant_matrix)
corr_mat.shape
```

```python
movies_names = rating_crosstab.columns
movies_list = list(movies_names)
```

```python
star_wars = movies_list.index('Star Wars (1977)')
```

```python
corr_star_wars = corr_mat[star_wars]
corr_star_wars.shape
```

```python
movie_names[(corr_star_wars < 1.0) & (corr_star_wars > 0.9.5)]
```
